```python
# Import the required modules
import xmltodict

# Open the file and read the contents
with open('Structural Steel.xml', 'r', encoding='utf-8') as file:
	my_xml = file.read()

# Use xmltodict to parse and convert the
# XML document
my_dict = xmltodict.parse(my_xml)

print(my_dict)
```

```python
import dicttoxml
xml = dicttoxml.dicttoxml(my_dict)
print(xml)
```
